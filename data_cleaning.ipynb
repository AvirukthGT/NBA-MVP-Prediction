```python
import pandas as pd
```

```python
mvps=pd.read_csv("mvps.csv",encoding="utf-8")
mvps['Player'] = mvps['Player'].apply(lambda x: x.encode('latin1').decode('utf-8'))
```

```python
mvps.tail(50)
```

```python
mvps=mvps[['Player','Year','Pts Won','Pts Max','Share']]
mvps.head()
```

```python
players=pd.read_csv("players.csv",encoding="utf-8")
players
```

```python
players.drop(columns=['Awards','Unnamed: 0','Rk'],inplace=True)
```

```python
players.head(20)
```

```python
players['Player']=players['Player'].str.replace('*','',regex=False)
```

```python
players[players['Player']=='James Harden']
```

```python
#dealing with multiple rows of same players playing with different teams in the same season
players.groupby(['Player',"Year"]).get_group(("Greg Anderson",1991))
```

```python
def single_row(df):
    combined_rows = df[df['Team'].isin(['2TM', '3TM'])]

    if not combined_rows.empty:
        # Use the combined row (e.g., full season), and replace team name with last actual team
        row = combined_rows.copy()
        row.loc[:, 'Team'] = df.iloc[-1]['Team']
        return row
    else:
        # No combined row, just take the last actual team played for
        return df.iloc[[-1]]  # keep as DataFrame

players=players.groupby(['Player',"Year"]).apply(single_row)
```

```python
players.index=players.index.droplevel()
```

```python
players.index=players.index.droplevel()
```

```python
players[players['Player']=='James Harden']
```

```python
combined=players.merge(mvps,how="outer",on=["Player","Year"])
```

```python
combined.tail(10)
```

```python
combined[['Pts Won','Share','Pts Max']]=combined[['Pts Won','Share','Pts Max']].fillna(0)
```

```python
teams=pd.read_csv("teams.csv")
```

```python
teams.head(30)
```

```python
teams=teams[~teams["W"].str.contains("Division")]
```

```python
teams['Team']=teams['Team'].str.replace("*",'')
```

```python
teams.head()
```

```python
teams["Team"].unique()
```

```python
combined['Team'].unique()
```

```python
combined=combined[combined["Team"]!='Team']
combined=combined[combined['Player']!='League Average']
```

```python
nicknames={}

with open("nicknames.txt") as f:
    lines=f.readlines()
    for line in lines[1:]:
        abbrev,name=line.replace("\n","").split(",")
        nicknames[abbrev]=name
```

```python
combined["Team"]=combined['Team'].map(nicknames)
```

```python
combined.head()
```

```python
stats=combined.merge(teams,how='outer',on=["Team","Year"])
```

```python
stats
```

```python
del stats['Unnamed: 0']
```

```python
stats=stats.apply(pd.to_numeric,errors="ignore")
```
